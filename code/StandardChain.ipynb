```python
from langchain_openai import ChatOpenAI
from langchain_anthropic import ChatAnthropic

import rich
from rich.console import Console
from rich.markdown import Markdown

console = Console()


gpt4 = ChatOpenAI(model="gpt-4o")
claud3 = ChatAnthropic(model='claude-3-opus-20240229')


from langchain_core.output_parsers import StrOutputParser
from langchain_core.prompts import ChatPromptTemplate

prompt = ChatPromptTemplate.from_template("Develop a business or product concept and around the following topic: {topic}. List dos and don'ts for the concept. Include a brief description of the concept and the target audience, how to price and market the business or product, and how to make it stand out from competitors. The output should be markdown.")
output_parser = StrOutputParser()

gpt_chain = prompt | gpt4 | output_parser
claud_chain = prompt | claud3 | output_parser
```

```python
gpt_output = gpt_chain.invoke({"topic": "Web scraping product prices"})
gpt_md = Markdown(gpt_output)
console.print(gpt_md)
```

```python
claud_output = claud_chain.invoke({"topic": "Web scraping product prices"})
claud_md = Markdown(claud_output)
console.print(claud_md)
```
